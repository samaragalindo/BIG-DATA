```python
# 1. Importar Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, confusion_matrix, classification_report

# Configuración de estilo
sns.set(style="whitegrid")

# 2. Carga de Datasets (Variables personalizadas: df_samara)
# Dataset Regresión: Medical Cost
df_samara_medico = pd.read_csv('insurance.csv')

# Dataset Clasificación: Heart Disease
df_samara_corazon = pd.read_csv('heart.csv')

# 3. Análisis de valores nulos y limpieza básica
print("--- Info Dataset Médico ---")
df_samara_medico.info()
print("\nNulos en Médico:\n", df_samara_medico.isnull().sum())

print("\n--- Info Dataset Corazón ---")
df_samara_corazon.info()
print("\nNulos en Corazón:\n", df_samara_corazon.isnull().sum())

# REQUISITO: Mostrar las primeras 5 filas
print("\n--- Head Dataset Médico ---")
display(df_samara_medico.head())
print("\n--- Head Dataset Corazón ---")
display(df_samara_corazon.head())
```

```python
# --- EDA para Regresión (Costos Médicos) ---
plt.figure(figsize=(10, 6))
# Seleccionamos solo numéricas para el mapa de calor
sns.heatmap(df_samara_medico.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación - Costos Médicos')
plt.show()

# Visualización: Fumar vs Costos
plt.figure(figsize=(6, 4))
sns.boxplot(x='smoker', y='charges', data=df_samara_medico)
plt.title('Impacto de Fumar en Costos Médicos')
plt.show()

# --- EDA para Clasificación (Enfermedad Cardíaca) ---
plt.figure(figsize=(6, 4))
sns.countplot(x='num', data=df_samara_corazon) # Corrected 'target' to 'num'
plt.title('Distribución de Enfermedad Cardíaca (0=No, >0=Sí)')
plt.show()
```

```python
# 1. Preprocesamiento (One-Hot Encoding)
df_samara_medico_encoded = pd.get_dummies(df_samara_medico, drop_first=True)

# 2. Definir X e y
X_med = df_samara_medico_encoded.drop('charges', axis=1)
y_med = df_samara_medico_encoded['charges']

# 3. División del dataset (Iniciales: SG)
X_train_SG, X_test_SG, y_train_SG, y_test_SG = train_test_split(X_med, y_med, test_size=0.2, random_state=42)

# 4. Entrenamiento del Modelo (Apellido: Galindo)
regresor_galindo = LinearRegression()
regresor_galindo.fit(X_train_SG, y_train_SG)

# 5. Predicción y Evaluación
y_pred_med = regresor_galindo.predict(X_test_SG)

r2 = r2_score(y_test_SG, y_pred_med)
rmse = np.sqrt(mean_squared_error(y_test_SG, y_pred_med))

print(f"R2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")

# 6. Interpretación de Coeficientes
coeficientes = pd.DataFrame(regresor_galindo.coef_, X_med.columns, columns=['Coeficiente'])
print("\nInfluencia de variables (Coeficientes):")
print(coeficientes.sort_values(by='Coeficiente', ascending=False))
```

```python
# 1. Preprocesamiento (One-Hot Encoding para el dataset de corazón)
# Primero, manejar los valores nulos para las columnas que tienen objetos y luego aplicar One-Hot Encoding

# Imputar valores nulos en columnas numéricas
df_samara_corazon.fillna(df_samara_corazon.mean(numeric_only=True), inplace=True)

# Imputar valores nulos en columnas categóricas con la moda
for column in df_samara_corazon.select_dtypes(include='object').columns:
    if df_samara_corazon[column].isnull().any():
        df_samara_corazon[column].fillna(df_samara_corazon[column].mode()[0], inplace=True)

# Ahora, aplicar One-Hot Encoding a todas las columnas categóricas
df_samara_corazon_encoded = pd.get_dummies(df_samara_corazon, drop_first=True)

# Definir X e y con el dataframe codificado
X_heart = df_samara_corazon_encoded.drop('num', axis=1)
y_heart = df_samara_corazon_encoded['num']

# 2. División del dataset (Iniciales: SG)
X_train_heart_SG, X_test_heart_SG, y_train_heart_SG, y_test_heart_SG = train_test_split(X_heart, y_heart, test_size=0.2, random_state=42)

# 3. Entrenamiento (Apellido: Galindo)
clasificador_galindo = LogisticRegression(max_iter=1000)
clasificador_galindo.fit(X_train_heart_SG, y_train_heart_SG)

# 4. Evaluación
y_pred_heart = clasificador_galindo.predict(X_test_heart_SG)

# Matriz de Confusión
cm = confusion_matrix(y_test_heart_SG, y_pred_heart)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - Clasificador Galindo')
plt.ylabel('Verdadero')
plt.xlabel('Predicho')
plt.show()

# Métricas
print("Accuracy:", accuracy_score(y_test_heart_SG, y_pred_heart))
print("\nReporte de Clasificación:\n")
print(classification_report(y_test_heart_SG, y_pred_heart))
```
